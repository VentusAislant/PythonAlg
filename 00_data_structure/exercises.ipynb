### 数组遍历

Q: 现有一个数组: `[1, 2, 3, 4, 5]`，利用两种方式遍历打印，包含元素和元素的索引
<details> 
<summary>点击展开答案</summary>
    
```python
arr = [1, 2, 3, 4]
# 索引方式遍历
for i in range(len(arr)):
    print(i, arr[i])

# 迭代器方式遍历
for i, val in enumerate(arr):
    print(i, val)
```

</details> 

````python
arr = [1, 2, 3, 4, 5]
````

在下面的单元格写入代码

````python
for idx, item in enumerate(arr):
    print(idx, item)

for idx in range(len(arr)):
    print(idx, arr[idx])
````
